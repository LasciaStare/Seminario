# Mortalidad por Cáncer de Estómago en Colombia (2008-2024)

## 1. Contexto del Proyecto
El cáncer de estómago representa una de las principales causas de mortalidad por neoplasias en Colombia, con una incidencia marcada por disparidades geográficas y socioeconómicas. Este estudio utiliza los microdatos de **Mortalidad No Fetal del DANE** para el periodo 2008-2024, con el fin de caracterizar el perfil epidemiológico de los fallecidos y preparar los datos para un modelo predictivo.

### Clasificación de Valores Faltantes (MCAR, MAR, MNAR)

#### **MNAR (Missing Not At Random - No aleatorio)**
*   **Estado_Civil_Madre (100%):** Es MNAR porque solo se diligencia si el fallecido es menor de un año. Esta base es de defunciones no fetales. Simplemente no lo capturan.
*   **Determinacion_Causa_A, B, C, D, E, F, G:** Son variables "indicadoras". El nulo significa **"No"**. Si `Determinacion_Causa_A` (Necropsia) es nulo, es porque NO se determino por necropsia.
*   **Municipio/Departamento_Ocurrencia_Hecho (99.9%):** Solo se llenan en casos de muerte violenta. Si la muerte es natural, estas columnas quedan vacías.

#### **MAR (Missing At Random - Aleatorio sistemático)**
*   **Codigo_EPS (9.32%):** Si el sitio de defunción fue fuera de la EPS. El nulo depende de la situación de la muerte.
*   **Hora_Defuncion (0.24%):** MAR. Si el cuerpo se encontró mucho después, la hora exacta es imposible de determinar.

#### **MCAR (Missing Completely At Random - Completamente aleatorio)**
*   **Departamento/Municipio_Residencia (0.06%):** Porcentajes tan bajos suelen ser errores humanos de digitación o fallos puntuales del sistema que no guardó el registro.

## 7) Distribucion y outliers en variables numericas

Se grafican histogramas y boxplots para variables clave, y se estima outliers con regla IQR.

```python
# Filtramos aquí porque si intentamos hacer "histogramas y boxplots"
# a categorías puras (`SEXO`, `GRU_ED1`) como intentábamos antes, Seaborn va a fallar.
# Solo dejamos variables de verdad continuas/cronológicas en un EDA cuantitativo.
vars_clave = [c for c in ["ANO", "MES", "HORA"] if c in eda_limpio.columns]

for c in vars_clave:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(eda_limpio[c].dropna(), kde=True, ax=axes[0], bins=20)
    axes[0].set_title(f"Histograma - {c}")
    sns.boxplot(x=eda_limpio[c], ax=axes[1])
    axes[1].set_title(f"Boxplot - {c}")
    plt.tight_layout()
    plt.show()

# Buscar Outliers numéricos (Usaremos las reglas de IQR en variables válidas)
outlier_rows = []
for c in num_cols_limpio:
    # A pesar de filtrar categoricas antes, no está mal asegurarse aquí
    if eda_limpio[c].dtype.name == "category" or eda_limpio[c].dtype == "object":
        continue
        
    s = eda_limpio[c].dropna()
    if s.nunique() < 5:
        continue
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1
    if iqr == 0:
        continue
    li, ls = q1 - 1.5 * iqr, q3 + 1.5 * iqr
    n_out = ((s < li) | (s > ls)).sum()
    outlier_rows.append({"columna": c, "outliers": int(n_out), "pct_outliers": 100 * n_out / len(s)})

outliers_df = pd.DataFrame(outlier_rows).sort_values("pct_outliers", ascending=False)
if len(outliers_df) > 0:
    display(outliers_df.head(20))
else:
    print("No se encontraron outliers significativos bajo la regla IQR en variables numéricas continuas.")
```

## 9) Relaciones entre variables (correlacion y cruces)
