# Tasa de Mortalidad por Cáncer de Estómago en Colombia (2008-2024)

## 1. Contexto del Proyecto
El cáncer de estómago representa una de las principales causas de mortalidad por neoplasias en Colombia, con una incidencia marcada por disparidades geográficas y socioeconómicas. Este estudio utiliza los microdatos de **Deuniones no Fetal del DANE** para el periodo 2008-2024, con el fin de caracterizar el perfil epidemiológico de los fallecidos y preparar los datos para un modelo predictivo.

### 3.1. Unificación de Causas de Muerte (2008-2018)

Debido al cambio de formato en los microdatos del DANE a partir de 2019, este bloque unifica las columnas históricas de causas directas, antecedentes y patológicas en una única variable compatible con el estándar moderno (formato IRIS).

**El proceso realiza:**
*   **Limpieza:** Normalización de códigos CIE-10 y eliminación de caracteres espurios.
*   **Deduplicación:** Eliminación de códigos redundantes manteniendo la jerarquía clínica.
*   **Estructuración:** Consolidación en una cadena unificada utilizando separadores técnicos (`/` para causas directas y `*` para patológicas).

```python
# Columnas del 2008 - 2018
cols_dir_ant =['C_DIR1', 'C_DIR12', 'C_ANT1', 'C_ANT12', 'C_ANT2', 'C_ANT22', 'C_ANT3', 'C_ANT32']
cols_pat = ['C_PAT1', 'C_PAT2']

def limpiar_codigo(codigo):
    """Limpia caracteres basura de los códigos CIE-10 del DANE."""
    if pd.isna(codigo): return None
    c = str(codigo).strip().replace('.', '') # Quita puntos
    # Lista negra de valores que no son códigos válidos
    if c in ('', 'nan', 'None', '"', '000', '0000'): 
        return None
    return c

def unificar_y_limpiar_dane(row):
    """
    Construye la variable CAUSA_MULT corrigiendo duplicados y saltos.
    """
    # 1. Extraer y limpiar directas/antecedentes
    causas_da =[limpiar_codigo(row[col]) for col in cols_dir_ant if col in row.index]
    causas_da = [c for c in causas_da if c is not None] # Filtrar nulos
    
    # 2. Extraer y limpiar patológicas
    causas_p = [limpiar_codigo(row[col]) for col in cols_pat if col in row.index]
    causas_p =[c for c in causas_p if c is not None] # Filtrar nulos

    # 3. ELIMINAR DUPLICADOS manteniendo el orden jerárquico
    # (Ej: Si el médico anotó C169 en Directa y Antecedente, solo queda la Directa)
    causas_da = list(dict.fromkeys(causas_da)) 
    
    # 4. Cruzar duplicados contra patológicas (Si una patológica ya está arriba, se quita)
    causas_p =[c for c in causas_p if c not in causas_da]
    causas_p = list(dict.fromkeys(causas_p))

    # 5. Formatear estilo IRIS (A/B/C*D)
    str_da = "/".join(causas_da)
    str_p = "/".join(causas_p)
    
    resultado = str_da
    if str_p:
        resultado = f"{resultado}*{str_p}" if resultado else f"*{str_p}"
        
    return resultado if resultado else np.nan
```

### 3.2. Consolidación de la Variable Maestra (2008-2024)

Este proceso integra los datos históricos procesados con los registros contemporáneos para generar la variable `CAUSA_MULT_FINAL`. Esta columna servirá como la fuente única de verdad para el análisis de causas múltiples en todo el rango temporal del estudio.

```python
# Aplicación al Dataset Completo
# 1. Crear la variable simulada perfectamente limpia para los años viejos
df['CAUSA_MULT_SIMULADA'] = df.apply(unificar_y_limpiar_dane, axis=1)

# 2. Crear la columna final combinando el formato nuevo (2019-2024) y nuestra simulada
df['CAUSA_MULT_FINAL'] = np.where(
    df['CAUSA_MULT'].notna() & (df['CAUSA_MULT'].astype(str).str.strip() != ''),
    df['CAUSA_MULT'],
    df['CAUSA_MULT_SIMULADA']
)

print("¡Consolidación finalizada con éxito! Duplicados eliminados en el texto consolidado.")
```

```python
# Filtramos los casos nulos
df_nulos = df[df['CAUSA_MULT_FINAL'].isna()]

# Miramos qué traían en las variables originales y si tienen C_BAS1
cols_ver =['ANO', 'C_DIR1', 'C_ANT1', 'CAUSA_MULT', 'C_BAS1']
# Filtramos solo las columnas que existan para no generar error
cols_ver =[c for c in cols_ver if c in df_nulos.columns]

print("=== MUESTRA DE LOS REGISTROS NULOS ===")
print(df_nulos[cols_ver].head(15).to_string(index=False))
```

### 3.3. Rescate de registros mediante Causa Básica

Se imputan los valores nulos en `CAUSA_MULT_FINAL` utilizando la información de `C_BAS1`. Este paso garantiza la **integridad total del dataset**, asegurando que cada registro cuente con al menos un diagnóstico codificado para el análisis.

```python
# Verificamos cuántos nulos hay antes
nulos_antes = df['CAUSA_MULT_FINAL'].isna().sum()
print(f"Nulos antes del rescate: {nulos_antes}")

# RESCATE: Si CAUSA_MULT_FINAL es nulo, toma el valor de C_BAS1
if 'C_BAS1' in df.columns:
    # Limpiamos los puntos del C_BAS1 por si acaso y lo pasamos a la final
    df['C_BAS1_LIMPIO'] = df['C_BAS1'].astype(str).str.strip().str.replace('.', '', regex=False)
    
    # Rellenamos los nulos
    df['CAUSA_MULT_FINAL'] = df['CAUSA_MULT_FINAL'].fillna(df['C_BAS1_LIMPIO'])
    
    # Opcional: Re-limpiar si algún C_BAS1 era 'nan' string
    df['CAUSA_MULT_FINAL'] = df['CAUSA_MULT_FINAL'].replace('nan', np.nan)

# Verificamos cuántos quedaron después
nulos_despues = df['CAUSA_MULT_FINAL'].isna().sum()
rescatados = nulos_antes - nulos_despues

print(f"Se lograron rescatar {rescatados} registros usando la Causa Básica (C_BAS1).")
print(f"Nulos definitivos (sin arreglo): {nulos_despues}")
```

### Jerarquización de la Calidad Diagnóstica

Se consolida una variable ordinal que clasifica el **grado de evidencia** detrás de cada certificado de defunción. La escala prioriza la confirmación anatomopatológica (necropsia) sobre la evidencia médica (historia clínica/pruebas) o el interrogatorio a familiares. 

Esta jerarquización permite ponderar la fiabilidad diagnóstica de los registros, identificando que el **80%** de los datos cuentan con un nivel alto de soporte clínico (niveles 1 y 2).

```python
import numpy as np

# 1. Definimos la jerarquía en orden de importancia (del más fuerte al más débil)
# 1 = Necropsia (A o E)
# 2 = Evidencia Médica / Clínica (B o C)
# 3 = Interrogatorio / Vigilancia (D o F)
# 4 = Sin Información (G o los que tienen 0.0 métodos)

condiciones = [
    (df['Determinacion_Causa_A'] == 1) | (df['Determinacion_Causa_E'] == 1), # Oro
    (df['Determinacion_Causa_B'] == 1) | (df['Determinacion_Causa_C'] == 1), # Plata
    (df['Determinacion_Causa_D'] == 1) | (df['Determinacion_Causa_F'] == 1), # Bronce
    (df['Determinacion_Causa_G'] == 1)                                      # Sin Info
]

valores = [1, 2, 3, 4]

# 2. Creamos la variable. El 'default=4' se encarga de los 11 registros que tienen 0.0
df['Calidad_Diagnostico'] = np.select(condiciones, valores, default=4)

# 3. Verificación final
print(df['Calidad_Diagnostico'].value_counts().sort_index())
```

### Transformación de Tipos y Optimización de Memoria

Tras la limpieza, el dataset aún conserva tipos de datos genéricos (`float64`) que consumen memoria innecesaria y dificultan la interpretación (ej. años con decimales). En esta fase, se realiza un proceso de **downcasting** y categorización para optimizar el rendimiento computacional.

```python
# 1. Definición CORREGIDA de grupos de variables
cols_categoricas = [
    'Departamento_Defuncion', 'Municipio_Defuncion', 'Area_Defuncion', 
    'Mes_Defuncion', 'Sexo', 'Grupo_Edad_Detallado', 'Estado_Civil', 
    'Departamento_Residencia', 'Municipio_Residencia', 'Sitio_Defuncion', 
    'Causa_Basica_CIE10', 'Certificador', 'Causa_Agrupada_Colombia', 
    'Grupo_Edad_General', 'Area_Residencia', 
    'Codigo_IPS', 'Nivel_Educativo', 'Pais_Residencia', 
    'Regimen_Salud', 'Asistencia_Medica', 'Accidente_Trabajo_Enf_Prof', 
    'Grupo_Etnico', 'Clase_EPS', 'Profesion_Certificador', 'Causa_666_667', 
    'Manera_Muerte_Consolidada', 'Entidad_Administradora_Salud'
]

cols_string = [
    'Nombre_IPS', 'Ocupacion_Habitual', 'Otro_Sitio_Defuncion', 
    'Causa_Multiple_CIE10'
]

# 2. Función de transformación CORREGIDA
def preparar_para_parquet(df_input):
    df = df_input.copy()
    
    # Manejo de Categorías (Códigos y Factores)
    for col in cols_categoricas:
        if col in df.columns:
            # Para códigos numéricos que deben ser categóricos:
            if df[col].dtype in ['float64', 'int64']:
                # Convertir a entero primero si es float sin decimales significativos
                # Luego a string para evitar el .0
                if df[col].dtype == 'float64' and (df[col] % 1 == 0).all():
                    df[col] = df[col].fillna(-1).astype('Int64').astype(str)
                else:
                    df[col] = df[col].astype(str)
            else:
                df[col] = df[col].astype(str)
            
            # Reemplazar valores nulos
            df[col] = df[col].replace(['nan', 'NaN', 'None', '-1', '-1.0', '<NA>'], 'SIN_DATO')
            # Convertir a categoría
            df[col] = df[col].astype('category')

    # Manejo de Strings (Texto abierto y Códigos Complejos)
    for col in cols_string:
        if col in df.columns:
            df[col] = df[col].astype("string").fillna("SIN_DATO")

    # Manejo de Numéricos Discretos (Downcasting optimizado)
    if 'Anio_Defuncion' in df.columns:
        # Años: 2008-2024 cabe en int16 (-32,768 a 32,767)
        df['Anio_Defuncion'] = pd.to_numeric(df['Anio_Defuncion'], errors='coerce').fillna(0).astype(np.int16)
        
    if 'Hora_Defuncion' in df.columns:
        # Horas: 0-23 cabe en int8
        df['Hora_Defuncion'] = pd.to_numeric(df['Hora_Defuncion'], errors='coerce').fillna(99).astype(np.int8)
        
    if 'Calidad_Diagnostico' in df.columns:
        # Valores típicamente 1-4, cabe en int8
        df['Calidad_Diagnostico'] = pd.to_numeric(df['Calidad_Diagnostico'], errors='coerce').fillna(0).astype(np.int8)
    
    # Optimización adicional: Convertir booleanos a int8 si existen
    # Por ejemplo: 'Accidente_Trabajo_Enf_Prof' podría ser binario
    
    return df

# 3. Aplicar optimización
# df = pd.read_csv('tu_archivo.csv', low_memory=False)

df_final = preparar_para_parquet(df_defunciones)

# 4. Verificar tipos después de la conversión
print("TIPOS OPTIMIZADOS:")
print(df_final.dtypes)
print("\nMEMORIA UTILIZADA:")
print(df_final.info(memory_usage='deep'))

# 5. Guardar en Parquet con compresión optimizada
nombre_archivo = "Datos/mortalidad_estomago_colombia_2008_2024.parquet"
df_final.to_parquet(nombre_archivo, engine='pyarrow', compression='snappy', index=False)

print(f"\nArchivo guardado exitosamente: {nombre_archivo}")

# 6. Comparación de tamaño
import os
tamaño = os.path.getsize(nombre_archivo) / (1024**2)  # En MB
print(f"Tamaño del archivo Parquet: {tamaño:.2f} MB")
```
